```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def brownian_motion(step, T):
    brownian = [0]
    x = [0]
    n = int(T/step)
    for i in range(n):
        brownian.append(brownian[i] + np.sqrt(step)*np.random.normal())
        x.append(x[i]+step)
    return x,brownian
```

```python
for i in range(10):
    x,W = brownian_motion(0.0001,1)
    plt.plot(x,W)
```

```python
def brownian_bridge(step,T=1,a=0,b=0):
    res = [0]
    x,B = brownian_motion(step,T)
    BT = B[-1]
    for i in range(1,len(x)):
        res.append(a*(T-x[i])/T + b*x[i]/T + B[i] - x[i]/T*BT)
    return x,res 
```

```python
for i in range(10):
    x,bridge = brownian_bridge(0.001,T=20,a=0,b=5)
    plt.plot(x,bridge)
```
